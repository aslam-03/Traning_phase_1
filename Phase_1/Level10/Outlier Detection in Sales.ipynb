```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

```python
data=pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv')
```

```python
data['Z_Score'] = zscore(data['Sales'])
outliers = data[data['Z_Score'].abs() > 3]
print(outliers)
```

```python
sns.boxplot(x=data['Sales'])
plt.title('Sales Distribution (Boxplot)')
plt.xlabel('Sales')
plt.show()
```
