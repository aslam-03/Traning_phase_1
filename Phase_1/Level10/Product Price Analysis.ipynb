```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Flipkart_Mobiles.csv')
data.head()
```

```python
range=data['Original Price'].max()-data['Original Price'].min()
variance=data['Original Price'].var()
standard_deviation=data['Original Price'].std
```

```python
print(f"Range:{range},Variance:{variance},Standard Deviation:{standard_deviation}")
```

```python

sns.histplot(data['Original Price'], kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()


sns.boxplot(x=data['Original Price'])
plt.title('Price Distribution (Boxplot)')
plt.xlabel('Price')
plt.show()
```
