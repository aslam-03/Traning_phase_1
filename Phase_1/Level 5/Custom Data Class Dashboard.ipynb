```python
import pandas as pd
import matplotlib.pyplot as plt
from dataclasses import dataclass

@dataclass
class DatasetStats:
    mean: float
    max: float
    min: float
    column: str

    def plot_histogram(self, df):
        """Plots a histogram of the specified column."""
        plt.hist(df[self.column], bins=20, color='skyblue', edgecolor='black')
        plt.xlabel(self.column)
        plt.ylabel("Frequency")
        plt.title(f"Histogram of {self.column}")
        plt.show()

def compute_stats(df, column):
    """Computes and returns dataset statistics as a data class instance."""
    return DatasetStats(
        mean=df[column].mean(),
        max=df[column].max(),
        min=df[column].min(),
        column=column
    )

def main():
   
    df = pd.read_csv("House_Price_dataset.csv")  
    
    print("Available columns:", df.columns.tolist())
    column = input("Enter the column name to analyze: ")

    if column not in df.columns:
        print("Invalid column name. Exiting.")
        return

    stats = compute_stats(df, column)

    while True:
        print("\nOptions: [1] Show Stats  [2] Plot Histogram  [3] Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            print(f"\nStatistics for {column}:")
            print(f"Mean: {stats.mean}")
            print(f"Max: {stats.max}")
            print(f"Min: {stats.min}")
        elif choice == "2":
            stats.plot_histogram(df)
        elif choice == "3":
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
