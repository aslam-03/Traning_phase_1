```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
df = pd.read_csv('Mall_Customers.csv')  
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
link_matrix = linkage(scaled_features, method='ward')
```

```python
plt.figure(figsize=(12, 6))
dendrogram(link_matrix)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.axhline(y=15, color='r', linestyle='--')  
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
labels = fcluster(link_matrix, t=3, criterion='maxclust')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(scaled_features[:, 1], scaled_features[:, 2], c=labels, cmap='Set2', s=60)
plt.xlabel('Annual Income (scaled)')
plt.ylabel('Spending Score (scaled)')
plt.title('Hierarchical Clustering - Customer Segments')
plt.grid(True)
plt.show()
    
```
